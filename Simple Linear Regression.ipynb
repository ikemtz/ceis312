## Load Training Dataset


```python
trainingArray = [
    [-7, -13],
    [-6, -11],
    [-5, -9],
    [-4, -7],
    [-3, -5],
    [-2, -3],
    [-1, -1],
    [0, 1],
    [1, 3],
    [2, 5],
    [3, 7],
    [4, 9],
    [5, 11],
    [6, 13],
    [7, 15]]
trainingData = pd.DataFrame(trainingArray, columns=['x', 'y'])
trainingData.head()
```

## Load Testing Dataset


```python
testArray = [
    [-10, -19],
    [-20, -39],
    [10, 21],
    [20, 41],
    [30, 61]]


testData = pd.DataFrame(testArray, columns=['x', 'y'])
testData.head()
```

## Popping the label series

Just check if label series exists before you do!


```python
if 'y' in trainingData.columns:
    trainingSeries = trainingData.pop('y')
    testingSeries = testData.pop('y')
```

## Creating the TensforFlow model

[sgd](https://realpython.com/gradient-descent-algorithm-python/#:~:text=Stochastic%20gradient%20descent%20is%20an,used%20in%20machine%20learning%20applications.) - Stochastic Gradient Descent

[mse](https://keras.io/api/losses/regression_losses/#meansquarederror-class) = Mean Squared Error


```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='sgd', loss='mse', metrics=['accuracy'])
```

## Training the Model


```python
model.fit(trainingData, trainingSeries, epochs=100)
```

## Evaluating performance of Model using TestData


```python
predictions = model.predict(testData)
```

```python
def evaluateModelPerformance(recordNum):
    print('** Evaluating Record Num: ', recordNum)
    print('     Value of x: ', testData.loc[recordNum]['x'])
    print('     Value of y: ', testingSeries.loc[recordNum])
    print('     Predicted Value of y: ', predictions[recordNum][0])
```

```python
evaluateModelPerformance(1)
evaluateModelPerformance(2)
evaluateModelPerformance(3)
```
